```python
#Visualising Image

def getImage(sample):
    img = sample.reshape((28,28))
    plt.imshow(img,cmap = 'gray')
    plt.show()

getImage(X_train[6])
print(Y_train[6])
```

```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))


def Knn(x,y,queryPoint,k = 5):
    m = x.shape[0]
    vals = []
    
    for i in range(m):
        d = dist(queryPoint,x[i])
        vals.append((d,y[i]))
    
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1] , return_counts = True)
    #print(new_vals)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
num = Knn(X_train,Y_train,X_test[108])
print(int(num))
```

```python
getImage(X_test[108])
```

```python
def getAccuracy(X_test, Y_test, X_train, Y_train):
    correct = 0
    length = len(X_test)
    for i in range(length):
        prediction = Knn(X_train , Y_train , X_test[i] )
        #print(prediction, correct)
        if Y_test[i] == prediction:
            correct += 1
            
    return (correct / float(length)) * 100.0   
    
```

```python
accuracy = getAccuracy(X_test, Y_test, X_train, Y_train)
print(accuracy)
```
